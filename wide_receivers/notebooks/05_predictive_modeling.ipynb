# Predictive Modeling: Can We Forecast NFL Success?

## Goal

After four notebooks of exploratory analysis—establishing NFL tiers (Notebook 01), analyzing college production vs. draft capital (Notebook 02), examining athletic testing's role (Notebook 03), and discovering YAC above expected as the key separator (Notebook 04)—it's time to ask a different question:

**Can we actually predict which college prospects will become Elite, Hit, Role, or Bust receivers in the NFL?**

This notebook will move beyond correlations and averages to build a simple classification model that combines everything we've learned:
- College production metrics (YPG, yards per catch, peak TDs)
- Combine athletic testing (40 time, 3-cone, broad jump)
- Conference pedigree
- Draft capital (round selected)

## Methodology

I'll use a straightforward approach that prioritizes interpretability over complexity:

1. **Data Preparation:** Clean and combine the master dataset from Notebook 04, handling missing values and encoding categorical variables
2. **Feature Selection:** Test which combination of pre-draft metrics most effectively predicts NFL tiers
3. **Model Training:** Start with logistic regression—a transparent model that shows us exactly which features matter most
4. **Evaluation:** Use confusion matrices to see where the model gets confused (e.g., mistaking Hits for Busts) and feature importance to understand what's driving predictions

## What I'm Trying to Uncover

- **The Holy Grail:** Is there a combination of metrics that consistently identifies Elite receivers?
- **Bust Warning Signs:** Do certain profiles (e.g., fast but unproductive) predict failure?
- **Market Inefficiencies:** Can we find receivers who outperform their draft slot based on underlying metrics?
- **Model Limitations:** Where does the model break down? This will tell us what we're still missing.

Let's find out if the numbers can actually tell us who will make it.
